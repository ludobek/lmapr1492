# Import et téléchargement de la structure (BaF2)

```python
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator
import matplotlib.pyplot as plt
#Téléchargement de la structure
with MPRester("hWWmIDodjhkgKVHj88m318vKs3nGtcQp") as m:
    structure = m.get_structure_by_material_id("mp-1029")
```

# Résultats des 3 premiers pics du diffractogrammes

```python
Lambda = 1.54060 #un rayonnement X incident Cu Kα de longueur d’onde λ=1.54060 Å.
pattern = XRDCalculator(Lambda).get_pattern(structure)
for i in range(0,3):
    theta = pattern.x[i]
    intensity = pattern.y[i]
    hkl = pattern.hkls[i][0].get("hkl")
    multiplicity = pattern.hkls[i][0].get("multiplicity")
    print("pic numéro", i+1, " pour un rayonnement X de longueur d’onde: λ=",Lambda, "Å :")
    print("  --> Angle 2Θ :", theta,"°")
    print("  --> Intensité du pic :", intensity)
    print("  --> Indices hkl :", str(hkl))
    print("  --> Multiplicité :" ,str(multiplicity))
    print("  --> d :", pattern.d_hkls[i], "\n")
```
